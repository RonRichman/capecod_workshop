```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from tryangle import Development, CapeCod, Chainladder
from tryangle.metrics import neg_cdr_scorer, neg_ave_scorer
from tryangle.model_selection import TriangleSplit
from tryangle.utils.datasets import load_sample
import pandas as pd
import numpy as np
```

# Import example dataset

```python
cas = pd.read_csv("cas train.csv")
```

```python
cas.head(20)
```

# Create and view Tryangle object

```python
X = load_sample("cas")
tscv = TriangleSplit(n_splits=5)
```

```python
X.triangle
```

# Basic Cape Cod example using Chainladder Package

```python
import chainladder as cl
```

```python
tri = cl.Triangle(cas, origin="origin", development="development", columns=["claim","premium"],cumulative=True)
```

```python
tri["claim"]
```

```python
tri['claim'].link_ratio.heatmap()
```

```python
dev = cl.Development(drop_high=False, drop_low=False).fit(tri['claim'])
```

```python
dev.ldf_
```

```python
transformed_triangle = cl.Development(drop_high=False, drop_low=False).fit_transform(tri['claim'])
```

```python
fit = cl.CapeCod(decay=0.8).fit(transformed_triangle, sample_weight=tri['premium'].latest_diagonal)
```

```python
fit.apriori_
```

```python
fit.ultimate_
```

```python
fit.ibnr_
```

# Optimising parameter selection using Tryangle Package

# Define parameter grid to optimise over

```python
param_grid = {
    "dev__n_periods": range(10,20,2),
    "dev__drop_high": [True,False],
    "dev__drop_low": [True,False],
    "cc__decay": [0,0.15,0.25,0.5,0.75,0.8,0.9,1]
}
```

# Define modelling pipeline

```python
pipe = Pipeline([("dev", Development()), ("cc", CapeCod())])
```

# Define optimisation parameters

```python
model = GridSearchCV(
    pipe, param_grid=param_grid, scoring=neg_ave_scorer, cv=tscv, verbose=1, n_jobs=8
)
```

# Run the optimisation process

```python
model.fit(X, X)
```

```python
model
```

# Optimisation Results

```python
model.cv_results_
```

```python
optimal_model = model.best_estimator_
```

```python
optimal_model
```

```python
model.best_params_
```

```python
model.best_score_
```
